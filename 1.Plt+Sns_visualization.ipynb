```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# sns 에서의 subplot

```python
# seaborn 에서의 subplot 은 이렇게 ax 를 지정해주어야한다
# 물론 plt.subplot(222) 를 쓴 이후에 sns 를 써도 되지만
# ax.sns 의 방법은 먹히지 않는다는 의미.
```

```python
data=pd.read_csv('C:/Users/Han/Desktop/Statistics/Python/Dataset/student-mat.csv')
data
fig, ax = plt.subplots(figsize=(10,4),ncols=3,nrows=1)
for i in range(1,4):
    sns.distplot(data['G'+str(i)],kde=True, rug=True,ax=ax[(i-1)]) # seaborn 에서의 subplot 은 이렇게 ax 를 지정해주어야한다
```

# sns 의 xlabel, title 등

```python
fake = pd.DataFrame({'cat': ['red', 'green', 'blue'], 'val': [1, 2, 3]})
fig = sns.barplot(x = 'val', y = 'cat', data = fake, color = 'black')
plt.xlabel("Colors") 
plt.ylabel("Values") 
plt.title("Colors vs Values") 
plt.show(fig)
```
