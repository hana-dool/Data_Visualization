# matplotlib

## 한글 나오게 하기

- matplotlib 에서 그래프를 그리다 보면, 한글폰트가 깨져나오는 경우가 있다.

```python
import numpy as np
data = np.random.randint(-100, 100, 50).cumsum()
data
plt.plot(range(50), data, 'r')
plt.title('시간별 가격 추이')
plt.ylabel('주식 가격')
plt.xlabel('시간(분)')
```

## dataframe 과 같이 나오게 하기

- 아래와 같이 dataframe 과 pandas 의 df 가 같이 나오게 하려면 display 메소드를 써야합니다.
- 그냥 df 를 써도, 나오긴 하지만 만일 그래프와 요약값이 함께 나오는 메서드를 정의하고 싶다면, diplay 와 함께 써야합니다.

```python
from IPython.display import display
```

```python
import numpy as np
data = np.random.randint(-100, 100, 50).cumsum()
display(pd.DataFrame(data.reshape(10,-1)))
plt.plot(range(50), data, 'r')
plt.show()
```
