# [1차원/범주형] pie 차트 시각화

```python
labels = ['frog', 'pig', 'dog', 'man']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0) # 어떤것을 다른것과 띄어서 볼 것인지. 이때 pig 가 띄어져있는것을 확인 가능하다.
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Pie Chart") 
plt.axis('equal')
plt.show()
```

# [2차원/수치형+수치형] histogram 2차원 시각화

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()

cb.set_label('counts in bin')
```

# [3차원/수치형 3개] 등고선으로 나타내고싶어

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)


n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
#입력 변수 x, y는 그대로 사용할 수 없고 meshgrid 명령으로 그리드 포인트 행렬을 만들어야 한다
# meshgrid 란?
# meshgrid(x,y)는 벡터 x 및 y에 포함된 좌표를 바탕으로 2차원 그리드 좌표를 반환합니다. 
# X는 각 행이 x의 복사본인 행렬이고, Y는 각 열이 y의 복사본인 행렬입니다.
# 좌표 X와 Y로 표현되는 그리드에는 length(y)개의 행과 length(x)개의 열이 있습니다.

ZZ = f(XX, YY)
plt.title("Contour plots")
plt.contourf(XX, YY, ZZ,levels=30, alpha=0.4, cmap='bwr')
# cmap,alpha 는 visual elementary에서.
# levels 는 몇개의 선을 정밀하게 만들것인가.
#plt.contour(XX, YY, ZZ) 그냥 이렇게 하면 선밖에 안나온다.
plt.colorbar()
# colorbar 를 형성한다.
plt.show()
```

# [3차원/수치형3개] 3-D PLOT

```python
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y) # 이역시 meshgrid 가 필요하다.

RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title("3D Surface Plot")
ax.plot_surface(XX, YY, ZZ, rstride=1, cstride=1, cmap='hot')
plt.show()
```

# [n차원/수치형 n 개] histo 겹쳐그리기

```python
# alpha 를 낮게 조절한다면, 다음과 같이 겹쳐서 그려질 수 있다.
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

plt.hist(x1, density=True,alpha=0.3 , label = 'korean')
plt.hist(x2, density=True,alpha=0.3 , label = 'japanese')
plt.hist(x3, density=True,alpha=0.3 , label = 'chinese')
plt.legend();
```
